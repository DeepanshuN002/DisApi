```python
df3=df.drop(['Disease'],axis=1)
df3.head()
```

```python
df3=pd.get_dummies(df3,prefix="")
```

```python
df3.head()
```

```python
df3.shape
```

```python
l=list(df3.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df3,y,test_size=0.2,random_state=50)
```

```python
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```
